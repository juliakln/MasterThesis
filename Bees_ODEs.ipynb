```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
          

# define system of ODEs with 2 reactions
def bees(y, t, k1, k2):
          bee, dead, p = y
          dydt = [-k1*bee*p, k1*bee*p, ((k1*bee*p) - (k2*p))]
          return dydt
          
# reaction rates
#k1 = 0.01
#k2 = 0.6
k1 = 1
k2 = 1

# initial conditions
#y0 = [100, 0, 2]
y0 = [1, 0, 1]

# solution for time interval between 0 and 100
#t = np.linspace(0,100,201)
t = np.linspace(0,20,51)

sol = odeint(bees, y0, t, args=(k1, k2))

fig = plt.figure()
plt.plot(t, sol[:, 0], 'g', label='Bees(t)')
plt.plot(t, sol[:, 1], 'r', label='DeadBees(t)')
plt.plot(t, sol[:, 2], 'b', label='Pheromone(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('E[x]')
plt.show()
#fig.savefig('figures/ode/ODE_1.png')

print(sol[-1])
```
